```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

# Trevelers

```python
ax =sns.countplot(data=df,x="BusinessTravel")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
dff=df[df["BusinessTravel"]=="Travel_Frequently"]
dff
```

# whose most travels

```python
ax =sns.countplot(data=dff,x="age_group")
for bars in ax.containers:
    ax.bar_label(bars)
```

# male or female

```python
ax =sns.countplot(data=dff,x="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

# male female in age group

```python
ax =sns.countplot(data=dff,x="age_group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

# educational field

```python
ax =sns.countplot(data=dff,x="EducationField")
sns.set(rc={"figure.figsize":(20,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# department belongs to

```python
ax =sns.countplot(data=dff,x="Department")
sns.set(rc={"figure.figsize":(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# marital status

```python
ax =sns.countplot(data=dff,x="MaritalStatus")
sns.set(rc={"figure.figsize":(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# most time travel in last year

```python
var =df.groupby(["age_group"],as_index=False)["TrainingTimesLastYear"].sum().sort_values(by="TrainingTimesLastYear",ascending=False)
sns.barplot(data=var,x="age_group",y="TrainingTimesLastYear")
```
